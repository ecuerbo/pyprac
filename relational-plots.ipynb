```python
import seaborn as sns
sns.set_style('ticks')

tips = sns.load_dataset('tips')

sns.relplot(x='total_bill',
           y='tip',
           hue='time',
           data=tips)
```

```python
import seaborn as sns

sns.set_style('dark')
tips = sns.load_dataset('tips')
markers = {'Lunch':"s", 'Dinner':'X'}

ax = sns.scatterplot(x='total_bill',
                    y='tip',
                    style='time',
                    markers=markers,
                    data=tips)
```

```python
import seaborn as sns

iris = sns.load_dataset('iris')

sns.scatterplot(x=iris.sepal_length,
               y=iris.sepal_width,
               hue = iris.species,
               style=iris.species)
```

```python
#Example 1: Basic visualization of fmri dataset using lineplot()
import seaborn as sns

sns.set_style('whitegrid')
fmri = sns.load_dataset('fmri')

sns.lineplot(x='timepoint',
            y='signal',
            data=fmri)

```

```python
#Ex.2: Grouping datapoints on the basis of category, here as region and event

import seaborn as sns

sns.set_style('whitegrid')
fmri = sns.load_dataset('fmri')

sns.lineplot(x='timepoint',
            y='signal',
            hue = 'region',
            style='event',
            data=fmri)
```
